```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df_movie=pd.read_csv("data/movie_comments.csv",dtype={"score":np.float64,"content":str})
df_cele=pd.read_csv("data/cele_comments.csv",dtype={"score":np.float64,"content":str})

group_id_movie=df_movie.groupby("work_id")
group_id_cele=df_cele.groupby("work_id")
```

```python
std_movie_score=[
    np.std(item[1]["score"]) for item in group_id_movie
]
std_cele_score=[
    np.std(item[1]["score"]) for item in group_id_cele
]

fig=plt.figure(figsize=(9,5))

y=std_movie_score
x=range(len(y))
ax=fig.add_axes([0,0,1,1])
ax.scatter(x,y,color="#54B345")


y=std_cele_score
x=np.linspace(0,250,80)
ax.scatter(x,y,color="#05B9E2")

ax.set_title("Contract of standard deviation")
ax.set_xlabel("Works index")
ax.set_ylabel("Standard deviation")
ax.set_xticklabels([])
ax.legend(["TOP250","Comparison"])

plt.show()
```

```python
std_movie_score=[
    (item[1]["score"]) for item in group_id_movie
]
std_cele_score=[
    (item[1]["score"]) for item in group_id_cele
]

plt.rcParams["font.size"]=18
fig=plt.figure(figsize=(20,15))

y=std_movie_score[200:250]
x=range(len(y))
ax=fig.add_axes([0,0.51,1,0.5])
ax.set_title("TOP250 box-plot")
ax.set_ylabel("Rating distribution")
ax.boxplot(y,sym="x",patch_artist=True,boxprops = {'color':'#F27970','facecolor':'#8983BF'})

y=std_cele_score[30:80]
ax=fig.add_axes([0,0,1,0.45])
ax.set_ylabel("Rating distribution")
ax.set_title("Comparison box-plot")
ax.boxplot(y,sym="x",patch_artist=True,boxprops = {'color':'#F27970','facecolor':'#8983BF'})


plt.show()
```
