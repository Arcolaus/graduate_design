```python
import pandas as pd  
import numpy as np
import matplotlib.pyplot as plt  
from scipy import stats

data_file=pd.read_csv("movie_comments.csv",dtype={"id":int,"score":np.float64})

id_group=data_file.groupby("id")

std_list=[]
mean_list=[]
std_div_mean=[]

for it,jt in id_group:
    std=np.std(jt["score"])
    mean=np.mean(jt["score"])
    std_list.append(std)
    mean_list.append(mean)
    std_div_mean.append(std/mean)



x1=range(len(std_list))
y1=std_list

fig1=plt.figure(figsize=(10,6))

ax1=fig1.add_axes([0,0,1,1])

ax1.set_xlabel("index")
ax1.set_ylabel("Standard Deviation")

ax1.scatter(x1,y1,alpha=0.6)

plt.show()
```

```python
x2=range(len(std_div_mean))
y2=std_div_mean

fig2=plt.figure(figsize=(10,6))

ax2=fig2.add_axes([0,0,1,1])
ax2.set_xlabel("index")
ax2.set_ylabel("ratio")

ax2.scatter(x2,y2,alpha=0.6)


plt.show()
```

```python
y=pd.read_csv("movie_detail.csv")["ratio"]
x=range(len(y))

fig=plt.figure(figsize=(10,6))

ax=fig.add_axes([0,0,1,1])
ax.set_xlabel("index")
ax.set_ylabel("ratio")

ax.scatter(x,y,alpha=0.6)

plt.show()
```
