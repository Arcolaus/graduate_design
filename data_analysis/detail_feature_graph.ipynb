```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

cele_detail=pd.read_csv("data/cele_detail.csv",on_bad_lines="skip")
movie_detail=pd.read_csv("data/movie_detail.csv",on_bad_lines="skip")

cele_ratio=cele_detail["ratio"]
movie_ratio=movie_detail["ratio"]

fig=plt.figure(figsize=(15,5))

ax=fig.add_axes([0, 0, 1, 1])
ax.set_ylim(0,0.7)

ax.scatter([i for i in range(movie_ratio.size)],movie_ratio,color="b",alpha=0.7)
ax.scatter(np.linspace(0,250,80),cele_ratio,color="r",alpha=0.7)

ax.set_ylabel("ratio")
# ax.set_xticks([])

ax.set_title("Comparison of ratio")
ax.legend(labels=("TOP250 Ratio","Celebrity Works Ratio"),loc="upper right")

plt.show()
```
